```python
#carichiamo i dati (che vengono da Matlab)
url = "https://raw.githubusercontent.com/massimocipressi/MLPNS_MCipressi/main/vis/dati.csv"
data = pd.read_csv(url,names=["Giorni_SIR","R","R_max","R_min","Giorni_uff","R_uff"])
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(data["Giorni_SIR"],data["R"],label="R(t) calcolato")
ax.fill_between(data["Giorni_SIR"],data["R_min"],data["R_max"],label="Incertezza: 1 sigma",alpha=0.5)
ax.plot(data["Giorni_uff"],data["R_uff"],label="R(t) ufficiale")
ax.plot([27,1093],[1,1],"r--",label="soglia")
ax.legend()
ax.set_title("Confronto fra dati ufficiali e calcolati con modello SIR")
ax.set_xlabel("Giorni: 1=28/1/2020")
ax.set_ylabel("R(t)")
ax.set_xlim(27,1093)
ax.set_ylim(-0.2,4.51)
```
